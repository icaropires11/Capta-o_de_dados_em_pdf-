## <font color=Yellow>Importando as Tabelas dos Pdfs

### Tabula, Try e while ...

```python
import tabula #Ferramenta para extrair dados de tabelas em arquivos PDF.


def extrair_tabela_do_pdf(caminho_pdf, pagina): #Define uma função chamada extrair_tabela_do_pdf que recebe dois parâmetros:

    try: #Inicia um bloco de código onde o programa tentará executar o código dentro do bloco e tratará possíveis exceções (erros)
        dados = tabula.read_pdf(caminho_pdf, pages=pagina)[0]
        return dados
    except Exception as e:
        print("Erro ao extrair tabela:", e)
        return None






contador = 8
while contador <= 13: #Loop que estipula que enquanto contador for menor que 13 vai rodar o código dentro da identação
        caminho_pdf = "copax1_" + str(contador) + ".pdf" # Caminho fica: copax1_8.pdf...copax1_9.pdf... copax1_10.pdf...
        pagina = 1    #Referente á pagina


        dados = extrair_tabela_do_pdf(caminho_pdf, pagina)  #Extrai a tabela do pdf (parametros:caminho do pdf, página)

        if dados is not None:       #Se os dados é diferente de nada ...
            caminho_csv = f"dados_copa_{contador}.csv"  # Defina um nome mais significativo para o arquivo CSV
            dados.to_csv(caminho_csv, index=False, sep=';')
            print(f"Tabela do arquivo {caminho_pdf} salva como CSV em {caminho_csv}.")
        else:
            print(f"Não foi possível extrair a tabela do arquivo {caminho_pdf}.")  # #Se os dados é não for diferente de nada ...

        contador += 1   #Enquanto contador for menor que 13 adiciona mais 1
```

## <font color = yellow>Tratamento de dados

```python
import pandas as pd
dados = pd.read_excel("Dados_Gerais.xlsx")
dados.columns = ["Retirar", "Pos", "Equipe", "Pontos", "Jogos", "Vitorias", "Empates", "Derrotas", "GP", "GC", "SG", "Porc" ]  #Renomeia as colunas
dados = dados.drop("Retirar", axis=1)  #Tirar coluna extra
```

###Tratando dados nulosn e Convertendo tipos de daos.

```python
dados.info()      #Exibir informações básicas do DataFrame
```

```python
# Tratamento de valores ausentes
# Preencher valores ausentes com zero
dados.fillna(0, inplace=True)
```

```python
dados["Pos"] = dados["Pos"].astype(int)      #Converter tipos de colunas
dados["Pontos"] = dados["Pontos"].astype(int)
dados["Jogos"] = dados["Jogos"].astype(int)
dados["Vitorias"] = dados["Vitorias"].astype(int)
dados["Empates"] = dados["Empates"].astype(int)
dados[ "Derrotas"] = dados["Derrotas"].astype(int)
dados["GP"] = dados["GP"].astype(int)
dados["GC"] = dados["GC"].astype(int)
dados["SG" ] = dados["SG"].astype(int)
dados[ "Porc" ] = dados[ "Porc"].astype(int)


#O erro ValueError indica que a conversão para o tipo int falhou porque uma das colunas contém valores não numéricos.
# O valor 'P' está causando o problema, já que não pode ser convertido para um número inteiro.
#Para lidar com isso, é necessário identificar quais são os valores não numéricos que estão causando o problema.
#Uma opção é imprimir os valores únicos na coluna problemática para entender quais são esses valores. Aqui está um exemplo:
```

###<font color=yellow>Categorizando os dados

```python
# Usar a função str.contains() para selecionar as linhas onde o valor da coluna "Equipe" contém "icaro" (ignorando maiúsculas/minúsculas)
Icaro = dados_tratados.query('Equipe.str.contains("icaro", case=False)', engine='python')
Jogos = Icaro['Jogos'].sum()
Vitorias = Icaro['Vitorias'].sum()
Empates = Icaro['Empates'].sum()
Derrotas = Icaro['Derrotas'].sum()
print(Jogos)
print(Vitorias)
print(Empates)
print(Derrotas)
```

```python
# Definindo a variável 'player' com o nome do jogador
player = "icaro"

Dados_Individuais = dados_tratados.query('Equipe.str.contains(@player, case=False)', engine='python')
# 'Equipe.str.contains(@player, case=False)': É a condição da consulta.
#...Neste caso, estamos usando a função str.contains() para verificar se o valor da coluna "Equipe" contém a string contida na variável player.
#O parâmetro case=False indica que a pesquisa deve ser feita ignorando maiúsculas/minúsculas, ou seja, a pesquisa é case-insensitive.
#engine='python': Esse parâmetro especifica o mecanismo de consulta a ser usado. Neste caso, estamos usando o mecanismo de consulta em Python.


# Calculando as somas
Jogos = Dados_Individuais['Jogos'].sum()
Vitorias = Dados_Individuais['Vitorias'].sum()
Empates = Dados_Individuais['Empates'].sum()
Derrotas = Dados_Individuais['Derrotas'].sum()

# Convertendo as variáveis numéricas em strings antes de concatená-las com outras strings
print("Número de jogos do " + player + " é " + str(Jogos))
print("Número de Vitórias do " + player + " é " + str(Vitorias))
print("Número de Empates do " + player + " é " + str(Empates))
print("Número de Derrotas do " + player + " é " + str(Derrotas))
```
